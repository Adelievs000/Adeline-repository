### Sintaks `tweepy` untuk mencari Tweet dari 7 hari terakhir.

Untuk mencari Tweet dari 7 hari terakhir, kita dapat menggunakan fungsi `search_recent_tweets` yang tersedia di Tweepy. kita harus memberikan permintaan **query** untuk menentukan data yang Anda cari. 


Secara default, Anda hanya akan mendapatkan ID Tweet dan teks Tweet untuk setiap Tweet. Jika Anda memerlukan bidang Tweet tambahan seperti `context_annotations`, `created_at` (waktu pembuatan Tweet), dll., kita dapat menentukan bidang tersebut menggunakan parameter `tweet_fields`, seperti yang ditunjukkan pada contoh di bawah ini.

Secara default, kita akan memperoleh 10 Tweet. Jika kita menginginkan lebih dari 10 Tweet per request, kita dapat menentukannya menggunakan parameter `max_results`. Maksimum Tweet per permintaan adalah `100`.

parameter `tweet_field` bisa diisi dengan field2 yang ada di link bawah ini:

https://developer.twitter.com/en/docs/twitter-api/data-dictionary/object-model/tweet

parameter `user_field` bisa diisi dengan field2 yang ada di link bawah ini:


https://developer.twitter.com/en/docs/twitter-api/data-dictionary/object-model/user

untuk parameter field lainnya dapat diakses pada link berikut:
https://docs.tweepy.org/en/v4.12.1/expansions_and_fields.html#tweet-fields-parameter


parameter `expansions` bisa diisi dengan field2 yang ada di link bawah ini:


https://docs.tweepy.org/en/v4.12.1/expansions_and_fields.html#expansions-parameter

```python
#Masukan Bearer Token untuk menggantikan 'Change this'
client = tweepy.Client(bearer_token='AAAAAAAAAAAAAAAAAAAAAABTlgEAAAAAVgb%2FUcE7bm3VCWyMWR54SKt2p7w%3DqeYH2OfjKyaiN5PiCa7ESxFiE1kJBD1ijZ7dQl0MbEkvPwIGdI')
```

```python
query = 'exo -is:retweet lang:id'
tweets = client.search_recent_tweets(query=query,
                                     tweet_fields=['context_annotations','public_metrics','created_at'],
                                     place_fields="geo",
                                     user_fields=["name"],
                                     expansions=['author_id','entities.mentions.username',
                                                 'geo.place_id','in_reply_to_user_id',
                                                 'referenced_tweets.id.author_id'
                                                 ],
                                     max_results=100)
```

```python
# Create a list of dictionaries, where each dictionary represents a single tweet
tweets_list = []
for tweet, user in zip(tweets.data, tweets.includes['users']):
    tweets_list.append({
        'created_at': tweet.created_at,
        'text': tweet.text,
        'public_metrics':tweet.public_metrics,
        'context_annotations':tweet.context_annotations,
        'name': user.name,
       'geo': tweet.geo
    })
```

```python
import pandas as pd
tweets_df = pd.DataFrame(tweets_list)
tweets_df.head()
```

### Mengonfigurasi Parameter Pencarian
Untuk menjalankan pencarian, Anda harus terlebih dahulu membuat objek konfigurasi.

```python
config = twint.Config()
config.Search = 'korupsi -is:retweet'
config.Limit = 30
```

Untuk memulai pencarian, jalankan kode berikut:

```python
twint.run.Search(config)
```

### Menyimpan Hasil ke dalam DataFrame "Pandas"

Anda juga perlu menambahkan variabel berikut ini ke dalam konfigurasi:

```python
config.Pandas = True
```
